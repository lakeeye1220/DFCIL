```python
import numpy as np
import matplotlib.pyplot as plt

ours =[
    [86.8,71.62,61.23,55.76,51.08],
[86.8,71.88,60.95,55.76,50.81],
[85.95,71.17,60.23,55.55,51.09],
]
```

```python
# create line graph using mean value and bar using std deviation measured by 3 runs
mean = np.mean(ours, axis=0)
max_var = np.max(ours, axis=0)
min_var = np.min(ours, axis=0)
# yerror = np.var(ours, axis=0)
yerror = np.abs(min_var-max_var)
plt.errorbar([1,2,3,4,5], mean, yerr=yerror, label='Ours')


plt.xlabel('Incremental Steps (T)')
plt.ylabel('Accuracy (\%)')
plt.title('Accuracy of different number of clusters')

```
